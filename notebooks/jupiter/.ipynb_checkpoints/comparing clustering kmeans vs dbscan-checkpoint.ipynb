```python
import numpy as np
import matplotlib.pyplot as plt
    
from sklearn import metrics
```

```python
np.random.seed(42)
plt.rcParams['figure.figsize'] = (30.0, 30.0)
plt.rcParams['figure.figsize'] = (30.0, 30.0)
plt.rcParams.update({'font.size': 36})
plt.rcParams['xtick.major.pad']='18'
plt.rcParams['ytick.major.pad']='18'
```

```python
d = np.random.uniform(0,5,     [1250,2])

#(x-x0)^2 + (y-y0)^2 < r^2
def circle(d, x0, y0, r):
    return (d[0]-x0)**2 + (d[1]-y0)**2 - r**2 < 0

def rectangle(d, x0, y0, x1, y1):
    return d[0]>x0 and d[0]<x1 and d[1]>y0 and d[1]<y1
```

```python
selection  = np.array([ rectangle(x, 1, 1, 1.5, 4.5) or 
                        rectangle(x, 1.5, 1.5, 4.5, 2) or
                        rectangle(x, 3, 3, 4, 4)
                        for x in d])

noise = np.random.uniform(0,5,     [20,2])

data = np.vstack((d[selection],  noise))
```

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, m_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
#x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
#y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1

x_min, x_max = 0.0 , 5.0
y_min, y_max = 0.0 , 5.0

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
meshdata = np.c_[xx.ravel(), yy.ravel()]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(init='k-means++', n_clusters=2, n_init=10)
kmeans.fit(data)
```

```python
# Obtain labels for each point in mesh.
Z = kmeans.predict(meshdata)

# Put the result into a color plot
Z = Z.reshape(xx.shape)
```

```python
plt.figure(1)
plt.clf()

plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower', alpha=0.1)

colors = plt.cm.Paired(np.linspace(0, 1, len(kmeans.cluster_centers_)))

reds  = kmeans.labels_==0
blues = kmeans.labels_==1

plt.plot(data[reds, 0],  data[reds, 1],  'ko', color=colors[0], markersize=25)
plt.plot(data[blues, 0], data[blues, 1], 'ko', color=colors[1], markersize=25)

# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=1500, linewidths=12,
            color='black', zorder=10)
plt.title('K-means clustering on two random 2d sets\n')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5, min_samples=5).fit(data)
```

```python
from sklearn import metrics

core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

print('Estimated number of clusters: %d' % n_clusters_)
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(data, labels))
```

```python
# Yellow used for outliers instead.
unique_labels = set(labels)
colors = np.array([[0.2, 0.2, 0.2, 1.0]])
colors = np.vstack(((plt.cm.Paired_r(np.linspace(0, 1, len(unique_labels)-1))), colors))

patches = []

fig, ax = plt.subplots()

for k, col in zip(unique_labels, colors):
    class_member_mask = (labels == k)

    xy = data[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=25)
        
    xy = data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=25)

plt.title('DBSCAN clustering on two random 2d sets + outliers\n')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
